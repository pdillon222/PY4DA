```python
import numpy as np
from numpy.random import randn
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Set controls within certain plots, pass in dictionaries
sns.distplot(dataset,bins=25,
            kde_kws={'color':'indianred','label':'KDE PLOT'},
            hist_kws={'color':'blue','label':'HIST'})
```

# Box and Violin Plots[51]

```python
#Another way of representing the probability density function using quartiles
```

url = 'http://en.wikipedia.org/wiki/Box_plot#mediaviewer/File:Boxplot_vs._PDF.svg'

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf,vert=False)
```

```python
#Violin plot= visual information of a boxplot with KDE information as well
#Create a normal distribution
data1 = stats.norm(0,5).rvs(100)
sns.set(style='ticks')
#Create two gamma distributions and concatenate
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])
#Box plot both data1 and data2
sns.boxplot(data=[data1,data2],whis=np.inf,vert=True)
```

```python
#Violing plot can be more revealing about data
sns.violinplot(data=[data1,data2])
```

```python
#Select different bandwidths
sns.violinplot(data=data2,bw=.01)
```

```python
sns.violinplot(data=data1,inner='stick')
```

# Regression Plots[52]

```python
#Visualising multiple regression with LM Plot [Seaborn comes with some data sets built in]
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#plot the dataset
sns.lmplot('total_bill','tip',tips)
```

```python
#We can use dictionaries to edit individual parts of the plot
#individually pass arguments
sns.lmplot('total_bill','tip',tips,
           scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
#multiple higher orders for quadratic fits
sns.lmplot('total_bill','tip',tips,order=4,
           scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
#Make a pure scatter plot with no regression
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips.head()
```

```python
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
tips.head()
```

```python
#Plot the new column, to see how it works with discrete variables
sns.lmplot('size','tip_pect',tips)
```

```python
#Add a jitter
```

url = 'http://en.wikipedia.org/wiki/Jitter'

```python
sns.lmplot('size','tip_pect',tips,x_jitter=.1)
```

```python
#Estimate the tendency of each bin, in the case above each bin is the size of the party
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
#Use hue subset to automatically define subsets along the column
#will automatically create hues for the sex column
#Leads to two regression plots, one on top of the other
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
#Does the day make a difference in how much the tip is?
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
#Seaborn supports local regression model fitting
```

url = 'http://en.wikipedia.org/wiki/Local_regression'

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True,line_kws={'color':'black'})
```

```python
#LM plot uses a lower level funvtion called regplot
#Regplot is actually the more robust plotting function
sns.regplot('total_bill','tip_pect',tips)
```

```python
#Regression plots using regplot can be added to other subplots
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)
sns.regplot('total_bill','tip_pect',tips,ax=axis1)
sns.violinplot(x='size',y='tip_pect',color=sns.palplot(sns.color_palette("BuGn_r")),data=tips,ax=axis2)
```

# Heatmaps and Clustered Matrices[53]

```python
#Import Seaborn dataset flights
flight_dframe = sns.load_dataset('flights')
flight_dframe.head()
```

```python
#Pivot the dataframe
#rows='month', columns='year', and the values we want = 'passengers
flight_dframe = flight_dframe.pivot('month','year','passengers')
flight_dframe
```

```python
#Let's make a heatmap
sns.heatmap(flight_dframe)
```

```python
#What if we want to know the visualization and the # of passengers?
sns.heatmap(flight_dframe,annot=True,fmt='d')
```

```python
#What if we want to choose a center for our color scheme
#Will create a diverging color scheme, white will be at the location we set
sns.heatmap(flight_dframe,center=flight_dframe.loc['January',1955])
```

```python
#Add heatmaps on axes of subplots
#make a subplot with 2 rows, 1 column
f,(axis1,axis2) = plt.subplots(2,1)
yearly_flights = flight_dframe.sum()
years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flights.values)
flights = pd.DataFrame(flights)

year_dframe = pd.concat((years,flights),axis=1)
year_dframe.columns = ['Year','Flights']

#Set a horizontal color bar
sns.barplot('Year',y='Flights',data=year_dframe,ax=axis1)
sns.heatmap(flight_dframe,cmap='Blues',ax=axis2,cbar_kws={'orientation':'horizontal'})
```

```python
#Use a cluster map, reformat the heatmap so similar rows are next to each other
#months will no longer be in order, but similar rows will be next to each other
sns.clustermap(flight_dframe)
```

```python
#Uncluster the columns
sns.clustermap(flight_dframe,col_cluster=False)
```

```python
#Set a standard scale, everything will be standardized by the year
sns.clustermap(flight_dframe,standard_scale=1)
```

```python
#Scale by rows
sns.clustermap(flight_dframe,standard_scale=0)
```

```python
#Standardize by the Z-score, subtract the mean and divide by stdev of each column
#rows will have a mean of 0 and a variance of 1
sns.clustermap(flight_dframe,z_score=1)
```
